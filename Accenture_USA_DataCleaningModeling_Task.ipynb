# Combined dataset

```python
print(df1.columns)
print(df2.columns)
print(df3.columns)
```

```python
df_2_3 = pd.merge(df2, df3, on='Type', how='inner')
print(df_2_3.columns)
df_1_3 = pd.merge(df1, df3, on='User ID', how='inner')
print(df_1_3.columns)
df_1_2_3  = pd.merge(df_2_3, df_1_3, on='User ID', how='inner')
print(df_1_2_3.columns)
```

```python
df_1_2_3.columns
```

```python
df_1_2_3 = df_1_2_3.drop(['Unnamed: 0_x_x','Unnamed: 0_y_x','Unnamed: 0_y_y','Unnamed: 0_x_y','URL','Datetime_x','Datetime_y'],axis = 1)
df_1_2_3.columns
```
